# Bài 1: Khởi tạo & Làm Âm Ảnh
1. Đọc ảnh mức xám
2. Tạo ảnh âm bản bằng công thức: $g = 255 - f$
3. Hiển thị ảnh gốc và ảnh âm bản

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

raw_img = cv2.imread("pictures/duy.jpg")

gray_img = cv2.cvtColor(raw_img, cv2.COLOR_BGR2GRAY)

neg_img = cv2.subtract(255,gray_img)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.title("Ảnh gốc")
plt.imshow(gray_img)
# plt.imshow(cv2.cvtColor(raw_img, cv2.COLOR_BGR2RGB))
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Ảnh âm bản")
plt.imshow(neg_img, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```
