```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Bc 1: c nh mc xm
img = cv2.imread('pictures/duy.jpg', cv2.IMREAD_GRAYSCALE)

# Bc 2: Tnh histogram gc
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

# Bc 3: Cn bng histogram
equalized_img = cv2.equalizeHist(img)

# Bc 4: Tnh histogram sau khi cn bng
hist_equalized = cv2.calcHist([equalized_img], [0], None, [256], [0, 256])

# Bc 5: Hin th nh gc v nh  cn bng
cv2.imshow('nh gc', img)
cv2.imshow('nh sau khi cn bng histogram', equalized_img)

# Bc 6: Hin th biu  histogram bng matplotlib
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Histogram nh gc')
plt.plot(hist_original, color='gray')
plt.xlim([0, 256])

plt.subplot(1, 2, 2)
plt.title('Histogram nh  cn bng')
plt.plot(hist_equalized, color='black')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
